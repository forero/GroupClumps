First plot to get an idea of the distribution of displacements between 
the most bound particles and the center of mass in DM group halos at $z=0.4$

All the selected halos have $300<V_{\rm rms}\ /\  {\mathrm{km\ s}}^{-1}<1000$

```python
halo_finder = "BDMW"
hubble = 0.73
filename="../data/Xoff_Rvir_snap66.dat"
halos_prop=loadtxt(filename,delimiter=",",skiprows=10, comments='#') 
```

```python
x_off = halos_prop[:,0] #dimensionless
r_vir = halos_prop[:,1] #in Mpc/h
x_off_dist = x_off * r_vir * 1000 / hubble
```

```python
#Figure with the separation distribution
fig = figure(1, figsize=(9.5,6.5))
ax = axes()
ax.set_xlabel("$\mathrm{Clump\ Offset [kpc]}$",fontsize=25)
ax.set_ylabel("$\mathrm{N_{halos}}$",fontsize=25)
ticklabels_x = ax.get_xticklabels()
ticklabels_y = ax.get_yticklabels()

for label_x in ticklabels_x:
    label_x.set_fontsize(18)
    label_x.set_family('serif')
for label_y in ticklabels_y:
    label_y.set_fontsize(18)
    label_y.set_family('serif')
n, bins, patches = ax.hist(x_off_dist, 20, normed=0, color = 'silver')
ax.set_xlim([0.0,200])
filename='offset_'+halo_finder

savefig(filename + '.eps',format = 'eps', transparent=True)
savefig(filename + '.pdf',format = 'pdf', transparent=True)
savefig(filename + '.png',format = 'png', transparent=True)
```

Here comes the fraction of halos with an offset larger than a given value

```python
x_off_dist_sort = sort(x_off_dist)
n_points = x_off_dist_sort.size
integrated = linspace(1.0,0.0, n_points)
```

```python
#Figure integrated offset distribution
fig = figure(1, figsize=(9.5,6.5))
ax = axes()
ax.set_xlabel("$\mathrm{Offset [kpc]}$",fontsize=25)
ax.set_ylabel("$\mathrm{F_{halos}(>Offset)}$",fontsize=25)
ticklabels_x = ax.get_xticklabels()
ticklabels_y = ax.get_yticklabels()

for label_x in ticklabels_x:
    label_x.set_fontsize(18)
    label_x.set_family('serif')
for label_y in ticklabels_y:
    label_y.set_fontsize(18)
    label_y.set_family('serif')
plot(x_off_dist_sort, integrated, color='black')
ax.set_xlim([0.0,200])
filename='integral_offset_'+halo_finder

savefig(filename + '.eps',format = 'eps', transparent=True)
savefig(filename + '.pdf',format = 'pdf', transparent=True)
savefig(filename + '.png',format = 'png', transparent=True)
```
